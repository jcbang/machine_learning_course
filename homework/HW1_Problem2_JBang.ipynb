```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```
# Our dataset
x1 = 2 * np.random.rand(200,1)
x2 = 2 * np.random.rand(200,1)
ys = 4 + (3 * x1 + 3 * x2) + np.random.rand(200,1)

# Our test dataset including labels
test_x1 = x1[100:]
test_x2 = x2[100:] 
test_ys = ys[100:]

# Our training dataset
train_x1 = x1[:100]
train_x2 = x2[:100]
train_ys = ys[:100]
```

```
# number of epochs
epochs = 10
# learning rate
lr = 0.01

w1 = np.random.randn(1)
w2 = np.random.randn(1)
b = np.zeros(1)

for epoch in np.arange(epochs):
  for i in np.arange(100):
    y_pred =(w1 * train_x1[i] + w2 * train_x2[i]) + b

    grad_w = (y_pred - train_ys[i]) 
    grad_w1 = grad_w * train_x1[i]
    grad_w2 = grad_w * train_x2[i]
    grad_b = grad_w

    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
    b  -= lr * grad_b
    
```

```
X1, X2 = np.meshgrid(test_x1, test_x2)
yHat = (w1 * X1 + w2 * X2) + b
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X1, X2, yHat, 30, cmap='inferno')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('yHat')
plt.show()
```
